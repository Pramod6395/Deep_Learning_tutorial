```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
(X_train,y_train) , (X_test,y_test) = keras.datasets.mnist.load_data()
```

```python
X_train_flattern =  X_train.reshape(len(X_train),28*28)
X_test_flattern =  X_test.reshape(len(X_test),28*28)
```

```python
print(X_train_flattern.shape)
print(X_test_flattern.shape)
```

```python
X_train_flattern[0]
```

```python
model= keras.Sequential([
    keras.layers.Dense(10,input_shape=(784,),activation='sigmoid')
])

model.compile(optimizer ='adam',
              loss = 'sparse_categorical_crossentropy',metrics=['accuracy'])

model.fit(X_train_flattern,y_train, epochs=5)
```

```python
# Lets scale our input array and see if get more accuracy

X_train_flattern_scale =  X_train_flattern/255
X_test_flattern_scale =  X_test_flattern/255
```

```python
model= keras.Sequential([
    keras.layers.Dense(10,input_shape=(784,),activation='sigmoid')
])

model.compile(optimizer ='adam',
              loss = 'sparse_categorical_crossentropy',metrics=['accuracy'])

model.fit(X_train_flattern_scale,y_train, epochs=5)
```

```python
model.evaluate(X_test_flattern_scale,y_test)
```

```python
plt.matshow(X_test[1])
```

```python
y_predicted= model.predict(X_test_flattern_scale)
y_predicted[1]
```

```python
np.argmax(y_predicted[1])
```

```python
y_predicted_labels = [ np.argmax(i) for i in y_predicted]
y_predicted_labels[:5]
```

```python
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)
cm
```

```python
import seaborn as sns
plt.figure(figsize=(15,7))
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
# Lets add hidden layear are check if we get performance improvement or not

model= keras.Sequential([
    keras.layers.Dense(100,input_shape=(784,),activation='relu'),
    keras.layers.Dense(10,activation='sigmoid')
])

model.compile(optimizer ='adam',
              loss = 'sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X_train_flattern_scale,y_train, epochs=5)
```

```python
# we can see after adding hidden layear accuracy jumps from 92 to 97%
model.evaluate(X_test_flattern_scale,y_test)
```

```python
y_predicted= model.predict(X_test_flattern_scale)
y_predicted_labels = [ np.argmax(i) for i in y_predicted]
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)

plt.figure(figsize=(15,7))
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
# instead of doing flatter manually keras have feature to flattern the array lets use that

model= keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(100 ,activation='relu'),
    keras.layers.Dense(10,activation='sigmoid')
])

model.compile(optimizer ='adam',
              loss = 'sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X_train,y_train, epochs=5)
```
